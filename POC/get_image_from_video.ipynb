```python
# Importing all necessary libraries 
import cv2 
import os 

# Read the video from specified path 
cam = cv2.VideoCapture("../Wideo/left_1.mov") 

try: 
	
	# creating a folder named data 
	if not os.path.exists('data'): 
		os.makedirs('data') 

# if not created then raise error 
except OSError: 
	print ('Error: Creating directory of data') 

# frame 
currentframe = 0

while(True): 
	
	# reading from frame 
	ret,frame = cam.read() 

	if ret: 
		# if video is still left continue creating images 
		name = '../Frames/Left/frame' + str(currentframe) + '.png'
		print ('Creating...' + name) 

		# writing the extracted images 
		cv2.imwrite(name, frame) 

		# increasing counter so that it will 
		# show how many frames are created 
		currentframe += 1
	else: 
		break

# Release all space and windows once done 
cam.release() 
cv2.destroyAllWindows() 
```

```python
import cv2
import os
from os.path import isfile, join
import re


def extract_number(filename):
        match = re.search(r'\d+', filename)
        return int(match.group()) if match else -1

def convert_frames_to_video(pathIn, pathOut, fps):
    frame_array = []
    files = [f for f in os.listdir(pathIn) if isfile(join(pathIn, f))]

    # Sort the files to ensure the correct order

    # Sort filenames based on the numeric part
    files = sorted(files, key=extract_number)

    for i in range(len(files)):
        filename = pathIn + files[i]
        print(f"Reading file: {filename}")
        img = cv2.imread(filename, cv2.IMREAD_UNCHANGED)
        if img is not None:
            height, width, layers = img.shape
            size = (width, height)
            frame_array.append(img)
        else:
            print(f"Error reading file: {filename}")

    print(f"Total frames to be processed: {len(frame_array)}")

    # Manually calculate the expected duration
    expected_duration = len(frame_array) / fps
    print(f"Expected video duration: {expected_duration} seconds")

    # Create VideoWriter with 'mp4v' codec and fps
    out = cv2.VideoWriter(pathOut, cv2.VideoWriter_fourcc(*'mp4v'), fps, size)

    for i in range(len(frame_array)):
        out.write(frame_array[i])

    out.release()

    cap = cv2.VideoCapture(pathOut)
    actual_duration = cap.get(cv2.CAP_PROP_FRAME_COUNT) / cap.get(cv2.CAP_PROP_FPS)
    print(f"Actual video duration: {actual_duration} seconds")
    cap.release()

def main():
    pathIn = '../Frames/Stitched/'
    pathOut = '../Wideo/output_video.mp4'
    fps = 60  # Keep the frame rate at 30 fps
    convert_frames_to_video(pathIn, pathOut, fps)

if __name__ == "__main__":
    main()



```

```python
from PIL import Image
import os

def resize_images(input_folder, output_folder, new_size):
    # Create the output folder if it doesn't exist
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    # Get a list of all files in the input folder
    files = os.listdir(input_folder)

    for file in files:
        # Check if the file is an image
        if file.lower().endswith(('.png', '.jpg', '.jpeg', '.gif', '.bmp')):
            input_path = os.path.join(input_folder, file)
            output_path = os.path.join(output_folder, file)

            # Open the image file
            with Image.open(input_path) as img:
                # Resize the image
                img = img.resize(new_size, Image.ANTIALIAS)

                # Save the resized image
                img.save(output_path)

if __name__ == "__main__":
    # Set the input and output folders and the new size
    input_folder = '../Frames/Stitched/'
    output_folder = '../Frames/Stitched/'
    new_size = (1853, 977)  # Set your desired width and height

    # Resize the images
    resize_images(input_folder, output_folder, new_size)

```
